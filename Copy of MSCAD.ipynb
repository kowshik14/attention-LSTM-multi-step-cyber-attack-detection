```python
import keras.backend as K
from keras.layers import Layer
from keras.layers import Input
```

```python
# Add attention layer to the deep learning network
class attention(Layer):
    def __init__(self,**kwargs):
        super(attention,self).__init__(**kwargs)
 
    def build(self,input_shape):
        self.W=self.add_weight(name='attention_weight', shape=(input_shape[-1],1), 
                               initializer='random_normal', trainable=True)
        self.b=self.add_weight(name='attention_bias', shape=(input_shape[1],1), 
                               initializer='zeros', trainable=True)        
        super(attention, self).build(input_shape)
 
    def call(self,x):
        # Alignment scores. Pass them through tanh function
        e = K.tanh(K.dot(x,self.W)+self.b)
        # Remove dimension of size 1
        e = K.squeeze(e, axis=-1)   
        # Compute the weights
        alpha = K.softmax(e)
        # Reshape to tensorFlow format
        alpha = K.expand_dims(alpha, axis=-1)
        # Compute the context vector
        context = x * alpha
        context = K.sum(context, axis=1)
        return context
```

```python
# reshape input to be [samples, time steps, features]
X_train = np.reshape(trainX, (trainX.shape[0], trainX.shape[1],1))
X_test=np.reshape(testX, (testX.shape[0], testX.shape[1],1))

print(X_train.shape)


batch_size = 32
```

```python
model = Sequential()
model.add(LSTM(64,return_sequences=True, input_shape=(66,1)))
model.add(attention())
model.add(Dense(32,activation='relu'))
model.add(Dense(6, activation="softmax"))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
checkpointer = callbacks.ModelCheckpoint(filepath="kddresults/nslkddmc/checkpoint-{epoch:02d}.hdf5", verbose=1, save_best_only=True, monitor='val_accuracy',mode='max')
csv_logger = CSVLogger('training_set_iranalysis.csv',separator=',', append=False)
m1=model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test,y_test), callbacks=[checkpointer,csv_logger])
```
